## 1Ô∏è‚É£ Importar Bibliotecas

```python

import pandas as pd
```

## 2Ô∏è‚É£ Criar Arquivo CSV de Entrada (Extra√ß√£o)

```python

data = {
    "Nome": ["Ana Souza", "Carlos Lima", "Mariana Alves"],
    "Conta": ["Conta Corrente Platinum", "Conta Digital", "Conta Universit√°ria"],
    "Cartao": ["Final 1234", "Final 5678", "Final 9012"]
}

df = pd.DataFrame(data)
df.to_csv("clientes.csv", index=False)
df
```

## 3Ô∏è‚É£ Leitura do Arquivo CSV

```python

df_clientes = pd.read_csv("clientes.csv")
df_clientes
```

## 4Ô∏è‚É£ Transforma√ß√£o ‚Äì IA Generativa (Simulada)

```python

def gerar_mensagem_marketing(nome, conta, cartao):
    return (
        f"Ol√° {nome}! üéâ Sua {conta}, com cart√£o {cartao}, "
        "oferece benef√≠cios exclusivos, praticidade no dia a dia "
        "e condi√ß√µes especiais pensadas especialmente para voc√™."
    )

df_clientes["Mensagem_Marketing"] = df_clientes.apply(
    lambda row: gerar_mensagem_marketing(
        row["Nome"], row["Conta"], row["Cartao"]
    ),
    axis=1
)

df_clientes
```

## 5Ô∏è‚É£ Carregamento ‚Äì Salvar Arquivo Final

```python

df_clientes.to_csv("mensagens_marketing.csv", index=False)
print("Arquivo mensagens_marketing.csv gerado com sucesso!")
```
